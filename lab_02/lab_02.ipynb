#### __Introducción__

```python
data=np.array([['Shirt','Medium','Blue'],
               ['Shirt','Large','Red'],['Pants','Medium','Khaki']])
data
```

```python
# transformar un dataframe a un array
data_df=pd.DataFrame(data,columns=['Product','Size','Color'])
data_df
```

```python
data_dic={'Product':['Shirt','Shirt','Pants'],'Size':['Medium','Large','Medium'],'Color':['Blue','Red','Khaki']}
#transformar un diccionario a un df

data_df=pd.DataFrame(data_dic)
data_df
```

```python
#añadir una columna
data_df['Price [$]']=[14,15,23]
data_df
```

```python
#añadir un registro
```

```python
data_df=data_df.append(pd.Series(['Skirt','Small','Pink',30],index=['Product','Size','Color','Price [$]']),ignore_index=True)
data_df
```

##### __Para mas información ver [click here](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html#appending-rows-to-a-dataframe)__

__Si deseamos añadir una columna aplicando una función a otra el metodo ideal a usar es aply, por ejemplo asumiendo que la columna Price no invluye el iva y deseamos tener otra fila que incluya este valor podemos utilizar__


```python
data_df['Price+IVA [$]']=data_df['Price [$]'].apply(lambda x: x*1.19)
data_df
```
